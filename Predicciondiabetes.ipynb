```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
diabetes.columns
```

```python
X=diabetes[['Embarazos', 'Glucosa', 'Presionarterial', 'EspesorPiel', 'Insulina','BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
X
```

```python
y=diabetes[['Resultado']]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=30)
```

```python
X_test
```

```python
y_test
```

```python
lrm=LinearRegression()
lrm.fit(X_train,y_train)
```

```python
prediccion = lrm.predict(X_test)
```

```python
prediccion
```

```python
y_test
```

```python
plt.scatter(y_test,prediccion)
```

```python
sns.distplot(y_test-prediccion)
```

```python
metrics.mean_absolute_error(y_test,prediccion)
```
