```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y,'green')
plt.title("Este es un grafico...")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'green')

plt.subplot(1,2,2)
plt.plot(y,x,'blue')
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'green')

plt.subplot(2,1,2)
plt.plot(y,x,'blue')
```

```python
figura=plt.figure()
grafico1=figura.add_axes([0.1,0.1,0.9,0.9])
grafico1.plot(x,y,'r')

grafico2=figura.add_axes([0.4,0.6,0.2,0.2])
grafico2.plot(x,y,'blue')
```

```python
#Multigraficos
figura, graficos = plt.subplots(nrows=2,ncols=2)
graficos[0][0].plot(x,y,'blue')
graficos[0][1].plot(x,y,'red')
graficos[1][0].plot(x,y,'green')
graficos[1][1].plot(x,y,'yellow')

graficos[0][0].set_title("Azul")
graficos[0][1].set_title("Rojo")
graficos[1][0].set_title("Verde")
graficos[1][1].set_title("Amarillo")
```

```python
figura=plt.figure(figsize=(1,2))
grafico = figura.add_axes ([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'red')
```

```python
figura=plt.figure()
graficos=figura.add_axes ([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2, label = "Al cuadrado")
graficos.plot(x,x**3, label = "Al cubo")
graficos.legend(loc=0)
```

```python
#color de graficos
figura=plt.figure()
graficos=figura.add_axes ([0,0,1,1])
graficos.plot(x,y,'#4386E1')

```

```python
#LINEAS DE GRAFICO
figura=plt.figure()
graficos=figura.add_axes ([0,0,1,1])
graficos.plot(x,y,color='red',linewidth=10,alpha=0.5,linestyle='--')
```
