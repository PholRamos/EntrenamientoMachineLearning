```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
casas.head()
```

```python
casas.columns
```

```python
X=casas[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms','Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
X
```

```python
y=casas[['Price']]
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=30)
```

```python
X_test
```

```python
y_test
```

```python
lrm=LinearRegression()
lrm.fit(X_train,y_train)
```

```python
from sklearn import metrics
```

```python
prediccion = lrm.predict(X_test)
```

```python
prediccion
```

```python
y_test
```

```python
plt.scatter(y_test,prediccion)
```

```python
sns.distplot(y_test-prediccion)
```

```python
#Metricas del error

#MAE (Mean absolute error)

metrics.mean_absolute_error(y_test,prediccion)
```

```python
#MSE (Media del error al cuadrado)

metrics.mean_squared_error (y_test,prediccion)
```

```python
#RMSE (raiz cuadrad del error)
np.sqrt(metrics.mean_squared_error (y_test,prediccion))
```

```python
#HACIENDO UNA PREDICCION

# Recolectar valores de características ingresados por teclado
avg_area_income = float(input("Ingresa el ingreso promedio del área: "))
avg_area_house_age = float(input("Ingresa la edad promedio de las casas en el área: "))
avg_area_num_rooms = float(input("Ingresa el número promedio de habitaciones en las casas del área: "))
avg_area_num_bedrooms = float(input("Ingresa el número promedio de dormitorios en las casas del área: "))
area_population = float(input("Ingresa la población del área: "))
```

```python
# Crear un DataFrame con los valores ingresados
nuevos_datos = pd.DataFrame({
    'Avg. Area Income': [avg_area_income],
    'Avg. Area House Age': [avg_area_house_age],
    'Avg. Area Number of Rooms': [avg_area_num_rooms],
    'Avg. Area Number of Bedrooms': [avg_area_num_bedrooms],
    'Area Population': [area_population]
})
```

```python
# Hacer la predicción con el modelo entrenado
prediccion = lrm.predict(nuevos_datos)
```

```python
# Mostrar la predicción
print("El precio predicho de la casa es:", prediccion[0])
```
